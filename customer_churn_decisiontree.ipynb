### 2.1 Entropy

```python
from math import log
import operator

# Calculate the entropy of the target attribute
def entropy(data, target_column):

    # Get the target column
    target = target_column

    # Count each class, how many of the target
    counts = data[target].value_counts()

    # The original total value
    total = len(data)

    # Define ent as 0 to change it later
    ent = 0

    # For each count,
    for count in counts:

      # Calculate the probability by dividing the count with total value.
      p = count / total

      # Use the formula log base 2
      # The negative sign (-=) is used because logging a fraction (probability)
      # results in a negative number, and we want a positive entropy value.
      ent -= p * math.log2(p)

    return ent
```

Explanation:  
== The code ==  
In the code, we calculate how many instances exist the target. For example in the question, we calculate how many "Yes" / 1 in the customer_churn. total = len(data) shows the total number of the data. For each count, we use count divided by total to calculate the probability to be used in calculating entropy. ent -= p * math.log2(p) is entropy formula to calculate the final result. Lastly, we return the final entropy.  
Entropy is the measure of uncertainty. In the decision tree, we need to split the data in a way that makes the result as "pure" as possible, means that we prefer less uncertainty and most data points belong to the same class.  
In the return result, if the ent is 0, it is pure and it is good result. If it is 0.5, some of the data are mixed. If the entropy value is 1, it is completely mixed.
We need to calculate entropy because it is useful in Decision Tree, it is a good measurement to help us to choose splits that reduce entropy.

### 2.2 Information Gain

```python
# Calculate the information gain
def information_gain(data, attribute, target_column):

    # Calculate the total entropy, it shows how the target is mixed before splitting
    total_entropy = entropy(data, target_column)

    # Initialise weighted_entropy
    weighted_entropy = 0

    # For each value of the chosen data
    for value in data[attribute].unique():

      # Create a subset where the attribute equals the current value
      subset = data[data[attribute] == value]

      # Calculate the weight
      weight = len(subset) / len(data)

      # Calculate the average entropy after splitting
      weighted_entropy += weight * entropy(subset, target_column)

    # Return the information gain
    # Information gain is the reduction in entropy
    return total_entropy - weighted_entropy
```

Explanation:  
== The code ==  
In the code, we calculate the entropy of the whole dataset. This tells us the uncertainty before splitting.   
Next, we initialise the weighted entropy as 0, the variable will store the entropy after splitting.   
for value in data[attribute].unique():, for each attribute, we don't repeat same value by using unique(). With this function, we can calculate entropy for each unique value. subset = data[data[attribute] == value] select rows where the attribute equals the value as subset. For example, all rows where Gender = Male form a subset. By dividing the length of subset (len(subset)) with total length (len(data)), we can get the weight. Then, we update the weighted_entropy by multiply the entropy of the subset and its weight. The weighted_entropy is the after split entropy.   
Lastly, return total_entropy (before split entropy) - weighted_entropy (after split entropy) to get the information gain.  
We need to use information gain as it helps us choose the best attribute to split and chooses the split that reduces uncertainty the most. Therefore, it can helps the decision tree to build a cleaner and more accurate tree.  

== The outcome ==  
No outcome is shown. The information_gain function is used to create the decision tree.  

### 2.3 Choose the best split

```python
# This function is used to choose the best attribute
def best_attribute(data, attributes, target_column):

    # Initialise an empty dictionary as gains to make comparison
    gains = {}

    # For each attribute given,
    for att in attributes:

      # Calculate its information gain
      gains[att] = information_gain(data, att, target_column)

    # Return the attribute with highest information gain
    return max(gains, key=gains.get)
```

Explanation:  
== The code ==  
This function is designed to identify the attribute that provides the best split for the dataset. Splitting is choosing one attribute and dividing the data based on its value, it is how a decision tree "ask questions" to reduce uncertainty, because all data is mixed together at the start, we need to use decision tree to split it. For example, gender has "Male" and "Female". We can split the attribute gender so we have two branches "Male" and "Female". After splitting, each group becomes more similar internally and the target label becomes more predictable. Continue to do splitting with other attributes to get the class (the result).   

To create a decision tree, we need to choose a best split. Best split is defined as the splitting that is low uncertainty. For instance one group mostly "Yes" and another group mostly "No", but not letting data mix together. By coding, we can define best split as the attribute with the highest Information Gain. As high Information Gain means that big reduction in entropy, the larger the difference (1 group 90%, another group 10%) shows the purer groups, therefore better predictions.   
The function starts from defining an empty dictionary. Then for each attribute, calculate their information gain and put into the dictionary. Lastly, return the maximum information gain within the gains dictionary. The returned attribute is the best split.  

== The outcome ==  
No outcome is shown. The best_attribute function is used to choose the best element to split in creating the decision tree.

### 2.4 Tree induction function

```python
def tree_induction(data, attributes, target_column, min_samples=200):

    # If all instances have same class,
    # cannot split further as no uncertainty left.
    if len(data[target_column].unique()) == 1:

      # If every row have the same label, return the first leaf node.
      return data[target_column].iloc[0]

    # Pre-pruning: stop if data too small
    # Cannot split further and must make decision
    if len(attributes) == 0 or len(data) < min_samples:

      # It will return the majority class (mode())
      return data[target_column].mode()[0]

    # Choose the best split by using the best_attribute function
    # created before
    best = best_attribute(data, attributes, target_column)

    # Create a dictionary to represent the node
    # Key: best attribute calculated above
    # Value: used to hold branches
    tree = {best: {}}

    # Loop over each possible value of the chosen attribute
    # Each value becomes a branch
    for value in data[best].unique():

      # Create subset
      subset = data[data[best] == value]

      # Handle empty subset
      if subset.empty:

        # Solve by predicting majority class of parent data (mode)
        tree[best][value] = data[target_column].mode()[0]

      # If the subset is not empty
      else:

        # if attribute best has already been used, cannot reuse
        remaining_attrs = [a for a in attributes if a != best]

        # Take a subset, remove used attribute and call build_tree again
        # Attach the returned subtree
        # Recursive function
        tree[best][value] = tree_induction(subset, remaining_attrs, target_column, min_samples)

    # Return the result
    return tree
```

Explanation:   
== The code ==   
This is the tree induction function. It recursively builds a decision tree by checking stop conditions, choosing the best attribute to split, splitting the data and repeating the process on each subset with the usage of entropy, information_gain and best_attribute functions created above.   
Step 1 and Step 2, we define the stopping condition. If all instances have the same class, we cannot split further as no uncertainty left as every row has the same label. Thus, we return the first label as all the result is same.  
If no attribute left or the attribute is smaller than the min_samples, we cannot split further to avoid overfitting but we need to decide the result immediately. Therefore we will use the mode() of the majority class. Both of them create leaf nodes but for different reasons.   
Step 3, we decide the best attribute to split the data in decision tree with the best_attribute function created just now.   
Step 4, we define a tree dictionary, and put the best as the key, the best will become the root of the tree to continue further split.   
Step 5, the for loop try every unique values one-by-one in each attribute, unique() means that we will not do the same value twice. For example if the support calls_bin value is 0, what is the target label, and when the value is 1, etc. They are the values.  
Step 6, we create subset to make it be tree branches.  
Step 7, For each data subset created by splitting on the selected attribute value, we first check whether the subset is empty. If the subset is empty, it means that there is no training instances corresponding to that attribute value. In this case, we will assign the majority class of the parent dataset as the result for the branch to avoid errors and make a reasonable predictions.  
If the subset is not empty, we remove the attribute that has been already used for splitting (best) from the attribute list to avoid redundancy in the decision tree. Then, we call the tree_induction function recursively on the subset with the remaining attributes to continue building the decision tree. This recursive process continues until one of the stopping conditions is met.  
Finally, this function will return the constructed tree that represented as a nested dictionary structure.

No outcome is shown as this is a model. The outcome will be shown if the model is used.

### 2.6 Classification function

```python
# The function is using the trained tree
def classify(tree, instance):

  # If the tree is not a dictionary,
  # it means we reached a leaf node, the result has come out
  # We return the tree with this stopping condition
  if not isinstance(tree, dict):
    return tree

  # Get the first key of the dictionary
  attribute = next(iter(tree))

  # Get the corresponding value by calling the
  # dictionary with its key in the test instance
  value = instance[attribute]

  # Check if the value is available in the tree
  if value in tree[attribute]:

    # Move down the tree to the next node
    # Recursive function
    # Repeat the function until meet the stop condition
    return classify(tree[attribute][value], instance)

  else:

    # If the value is missing, return None
    return None
```

Explanation:  
== The code ==  
Classification is performed by traversing the decision tree from the root to a leaf node based on the attribute values of the input instance.  
The classification function is using the trained tree.  
Step 1 is the stopping condition. If the tree is not dictionary, means that we reach the leaf node, the result has come out, because the tree is build using dictionaries and the leaf nodes are class labels (the result). Therefore, we can return the leaf node (result) as the function is done.  
Step 2: get the attribute used at this node. It is the first key of the dictionary.  
Step 3: look up the value of that attribute in the dictionary instance.  
Step 4: find if the value is available in the tree dictionary value. If yes, call the classification function recursively to classify the data until it reached a leaf node. If the value is missing, return None. This is another stop condition.  

== The outcome ==   
No outcome is shown. The classification function is used to use the train data to do classification.  
If the classification function is used, it will return the class label (value) such as 1.0 or 0.0 as the final prediction.  

The Decision Tree is constructed using a recursive tree induction algorithm. At each node, entropy and information gain are calculated for all candidate attributes. The attribute with the highest information gain is selected to split the data. This process repeats for each branch to create subtrees until a stopping condition is met (arrive the leaf node).

### 2.7 Predict function to test data

```python
def predict(data, tree):
    # Convert DataFrame to dictionary
    instances = data.to_dict(orient='records')

    # Use classification function to use the trained model
    # Classify the model with each instance
    predictions = [classify(tree, inst) for inst in instances]

    # Put the result into predictions list
    data["Predicted_Churn"] = predictions

    return data["Predicted_Churn"]
```

From the data, we convert each row of record into dictionary and define it as instances.  
For each instance, use classify() to predict the result with tree_induction function.  
Put the result into the list named predictions.

### 3.2 Pre-pruning Explanation

Pre-pruning was applied by introducing a minimum sample threshold(min_samples) in tree_induction function.   
if len(attributes) == 0 or len(data) < min_samples:  
    return data[target_column].mode()[0]  
  If the number of instances at a node is below this threshold (min_samples), the splitting process stops and the majority class is returned. Otherwise, it will continue splitting until every single leaf is "pure" (Entropy = 0). Pre-pruning reduces overfitting noise data, while keeping the implementation simple and stable decision tree. Besides, it can effectively preserve computationally efficient.
